```python
families = {
    "Lester-Boal": ["Paul", "Sarahlynn", "Ellie", "Ada", "Teddy"],
    "Duck": ["Donald", "Daisy"],
    "Kardashian": ["Kim", "Chloe", "Courtney", "Rob"]
}
```

```python
for last_name, first_names in families.items():
    for name in first_names:
        print(name, last_name)
```

```python
## converting dict to list
names = []
for last_name, first_name in families.items():
    for name in first_names:
        names.append([name, last_name])
```

```python
names
```

```python
[[name[1], name[2]] for name in names if name[0] == 'Paul']
```
