```python
dfs = pd.read_html('https://emedicine.medscape.com/article/1947145-workup#c13')
```

```python
TNM_staging = dfs[5].fillna(method = 'ffill')
```

```python
header = TNM_staging.iloc[0]
TNM = TNM_staging[1:] #take the data less the header row
TNM.columns = header
```

```python
TNM_data = TNM.to_json(orient = 'records')
```

```python
def TNM_stage(tumor, node, metastasis):
    with open(r"res.json") as af:
        obj = json.load(af)
   
    for a in obj:
        if a['T'] == tumor and a['N']== node and a['M']== metastasis:
            stage = (a['Stage'])
            return stage
    
```

```python
TNM_stage('T0', 'N1', 'M0')
```

```python
import ast
res = ast.literal_eval(TNM_data)
res
```

```python
with open("res.json", "w") as outfile:
    json.dump(res, outfile)
```

```python
def cancer_treatment(stage, er, pr, her2):
    with open(r"cancer_drugs.json") as f:
        obj = json.load(f)
    for i in obj:
        if i['STAGE'] == stage and i['ER']== er and i['PR']== pr and i['HER2']== her2 :
            treatment = (i['TREATMENT']['DRUG'])
            return treatment
```

```python
cancer_treatment('Stage 2', 'ER Negative', 'PR Negative', 'HER2-Positive')
```
