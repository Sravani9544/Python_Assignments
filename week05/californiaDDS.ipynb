```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Data from https://www.kaggle.com/wduckett/californiaddsexpenditures?select=californiaDDSDataV2.csv
```

```python
california = pd.read_csv('/data/californiaDDSDataV2.csv')
```

```python
california
```

```python
sns.boxplot(x='Ethnicity', y='Expenditures', data=california)
```

```python
sns.catplot(x='Ethnicity', y='Expenditures', data=california, aspect=30/8)
```

```python
sns.catplot(x='Ethnicity', y='Expenditures', data=california, aspect=30/8, hue='Age Cohort')
```

```python
pd.options.display.float_format = '{:,.0f}'.format
california.pivot_table(index='Age Cohort', columns='Ethnicity', aggfunc='mean', values='Expenditures')
```

```python
sns.violinplot(california, x='Ethnicity', y='Age')
```

```python
california.pivot_table(index='Age Cohort', columns='Ethnicity', aggfunc='count', values='Expenditures')
```

```python
sns.pairplot(california)
```
